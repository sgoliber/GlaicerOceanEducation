## Module 1: What happens if we have a constant temperature change?

In this module we will explore how increasing the rate of ocean melt will impact the location of our glacier.

To use this part of the module, select your **Ocean Melt change** and **Animation speed**.

| **Ocean Melt** | **Rate Description** | **Temperatue** |
|----------------|-----------------|----------------|
| 50% increase      | Moderate increase   | Warm           |
| 75% increase        | High increase       | Warmer         |
| 100% increase        | Highest increase    | Warmest        |

```python
ocean_step_ui = ui.Togglebuttons(
            name='Ocean Step Change',
            description="",
            value=50,
            options=[50,75,100]
        )

gifspeed_step_ui = ui.Togglebuttons(
            name='Animation Speed',
            description="",
            value='Normal',
            options=['Slow','Normal','Fast']
        )

ui.Form([ocean_step_ui,gifspeed_step_ui],name='Options')
```

```python
def display_oceanstep(s):
    out_step.clear_output()
    ocs = str(ocean_step_ui.value)
    speed = {'Fast': '8', 'Normal': '4', 'Slow': '2'}
    fps = speed.get(gifspeed_step_ui.value)
    with out_step:
        display(Image(filename='gifs/test/bumpy_ocnstep/bumpy_ocnstep_'+ocs+'pct/'+str(fps)+'.gif'))
```

```python
ui.Submit(label='Run', show_progress = True, start_func = display_oceanstep)
```

```python
out_step = widgets.Output(layout={'border': '1px solid black'})
out_step.layout.width = '800px'
display(out_step)
with out_step:
    ui.Submit(label='Run', show_progress = True, start_func = display_oceanstep)
```

## Module 2: What happens if we change the temperature over time?

In this module we will explore how increasing the amount of ocean melt will impact the location of our glacier. To use this part of the module, select your **Animation speed** and the ocean melt will change on it's own.


| **Ocean Melt** | **Description** | **Temperatue** |
|----------------|-----------------|----------------|
| 25 m/yr        | Moderate melt   | Warm           |
| 45 m/yr        | High Melt       | Warmer         |


```python
gifspeed_sqr_ui = ui.Togglebuttons(
            name='Animation Speed',
            description="",
            value='Normal',
            options=['Slow','Normal','Fast']
        )

ui.Form([gifspeed_sqr_ui],name='Options')
```

```python
def display_oceansqr(s):
    out_sqr.clear_output()
    speed = {'Fast': '8', 'Normal': '4', 'Slow': '2'}
    fps = speed.get(gifspeed_sqr_ui.value)
    with out_sqr:
        display(Image(filename='gifs/test/bumpy_ocnsq/bumpy_ocnsq_50pct/'+str(fps)+'.gif'))
        
        
```

```python
ui.Submit(label='Run', show_progress = True, start_func = display_oceansqr)
```

```python
out_sqr = widgets.Output(layout={'border': '1px solid black'})
out_sqr.layout.width = '800px'
display(out_sqr)
with out_sqr:
    ui.Submit(label='Run', show_progress = True, start_func = display_oceansqr)
```
